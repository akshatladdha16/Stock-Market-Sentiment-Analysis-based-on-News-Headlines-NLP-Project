Data set in consideration is a combination of the world news and stock price shifts. data ranges fromm 2000 to 2016 and 2000 to 2008 done by web scarpping in yahoo finance. Class 1: stock price increased and 0 denotes same or decreased

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Data.csv')
df.head()
```

```python
#dividing in train and test dataset based on year
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']
```

```python
# removing punctuation and other nonalphanumeric things
#feature engineering
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
data.head()
```

A RegEx is a powerful tool for matching text, based on a pre-defined pattern. It can detect the presence or absence of a text by matching it with a particular pattern, and also can split a pattern into one or more sub-patterns. The Python standard library provides a re module for regular expressions. Its primary function is to offer a search, where it takes a regular expression and a string. Here, it either returns the first match or else none.

```python
#renaming column names
list1=[i for i in range(25)]
new_index=[str(i) for i in list1] #new_index is a string of numbers 1 to 25
data.columns=new_index
```

```python
#converting in lower case
for index in new_index:
    data[index]=data[index].str.lower()
data.head()
# important for tfidf BOW and word2vec coz a and A is different in these vectorizers
```

```python
# for application of NLP we have to merge all headlines in one para
headlines=[]
for row in range(0,len(data.index)): # have to go by column
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
headlines[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer #BOW
from sklearn.ensemble import RandomForestClassifier 
```

```python
#implementaion of BOW 
countvector=CountVectorizer(ngram_range=(2,2)) # have to try again n again with different numbers
traindataset=countvector.fit_transform(headlines)
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
#predicting for test data
test.head()
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
testdataset=countvector.transform(test_transform)
pred=randomclassifier.predict(testdataset)
```

```python
# importing library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],pred)
print(matrix)
```

```python
print(classification_report(test['Label'],pred))
```

```python
print(accuracy_score(test['Label'],pred))
```
